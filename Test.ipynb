```python
import xml.etree.ElementTree as ET
 
tree = ET.parse('../../data/Phamacity/linhdam2/linhdam2_10.xml')

root = tree.getroot()
cnt_x = float(root[6][5][0].text)
cnt_y = float(root[6][5][1].text)
w = float(root[6][5][2].text)
h = float(root[6][5][3].text)
angle = float(root[6][5][4].text) *180/np.pi
rotated_box = (cnt_x, cnt_y, w, h, angle)
```
